## **Problem 11: String Methods**
- Write a Python program to check if all elements of a string are alphabetic (حروف أبجدية فقط).

```python
s1 = 'MyNameIsAli'
s2 = 'My Name IsAli'

# Answer here

if s1.isalpha():
    print("All characters in s1 are alphabetic")
else:
    print("s1 characters are non-alphabetic")

if s2.isalpha():
    print("All characters in s2 are alphabetic")
else:
    print("s2 characters are non-alphabetic")
```

- Write a Python program to check if all elements of a string are numbers.

```python
s1 = '123456789'
s2 = '125.5'

# Answer here

if s1.isdigit():
    print("All characters in s1 are digits")
else:
    print("s1 contains non-digit characters")

if s2.isdigit():
    print("All characters in s2 are digits")
else:
    print("s2 contains non-digit characters")

```

## **Problem 23: IF Condition**
#### Write a program that takes a user's age as input and determines their life stage.
* Print "Child" if the age is less than 13,
* "Teen" if the age is between 13 and 19,
* "Young Adult" if the age is between 20 and 35,
* and "Adult" if the age is 36 or older.

```python
# Answer here

age = int (input("Enter your age : "))

if age < 13 :
    print("Child")
elif 13 <= age <= 19 :
    print("Teen")
elif 20 <= age <= 35 :
    print("Young Adult")

else :
    print("Adult")
```
