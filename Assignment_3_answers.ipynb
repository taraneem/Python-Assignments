## <p style="color:orange;">Q.19 Write a Python program to create an intersection of sets.</p>
> Explanation : In mathematics, the intersection of two sets A and B, denoted by A ‚à© B, is the set containing all elements of A that also belong to B (or equivalently, all elements of B that also belong to A).

```python
# Answer here
set_A = {1,2,3,4,5}
set_B = {4,5,6,7,8}
intersection = set_B.intersection(set_A)
print(f'Intersection of set_A and set_B: {intersection}')
```

## <p style="color:orange;">Q.20 Write a Python program to create a union of sets.</p>
> Explanation :In mathematics, The union of two sets A and B is the set of elements which are in A, in B, or in both A and B.

```python
# Answer here
set_A = {1,2,3,4,5}
set_B = {4,5,6,7,8}
union = set_A.union(set_B)
print(f'Union of set_A and set_B: {union}')
```

- Write a for loop to iterate from 0 to 100 and print only even numbers

```python
# Answer here
even_numbers = []
for i in range(101):
    if i % 2 == 0:
        even_numbers.append(i)

print(f'Even numbers from 0 to 100: {even_numbers}')
```

- Write a for loop to iterate from 0 to 100 and print only odd numbers

```python
# Answer here
odd_numbers = []
for i in range(101):
    if i % 2 != 0:
        odd_numbers.append(i)

print(f'Odd numbers from 0 to 100: {odd_numbers}')
```

- Write a for loop that prints numbers from 10 to -10 and break the loop when the number is negative

```python
# Answer here
for i in range(10 , -11 , -1) :
    if i < 0 :
        break
    print(i)
```

- Write a for loop that prints numbers from 10 to -10 and skip the iteration where number is zero

```python
# Answer here

for i in range(10 , -11 ,-1) :
    if i != 0:
        print(i)
    else :
        continue
```
